```python
list = [1,3,4,6,6,6,6,7,8,10,14]

def bisect_left(nums, x):
    l,r = -1,len(nums)

    while l < r - 1:
        mid = (l + r) // 2
        if nums[mid] < x:
            l = mid
        else:
            r = mid
        
    return r                    # 最终空位就是插入点

print(bisect_left(list,6))
```
